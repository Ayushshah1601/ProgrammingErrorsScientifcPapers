```python
def download_pdf(doi, file_name):
    base_url = "https://doi.org/"  # DOI resolver
    doi_url = base_url + doi

    try:
        # Get the redirect URL
        response = requests.get(doi_url, allow_redirects=True)
        response.raise_for_status()  # Raise error for HTTP issues
        final_url = response.url  # Get the resolved page URL

        # Fetch the final webpage
        page = requests.get(final_url)
        page.raise_for_status()

        # Parse HTML
        soup = BeautifulSoup(page.text, "html.parser")

        # Find PDF links
        pdf_link = None
        for link in soup.find_all("a", href=True):
            href = link["href"]
            if "pdf" in href.lower():
                pdf_link = urljoin(final_url, href)
                break  # Stop after finding the first PDF

        if pdf_link:
            pdf_response = requests.get(pdf_link, stream=True)
            pdf_response.raise_for_status()

            # Save PDF
            pdf_path = os.path.join(pdf_dir, f"{file_name}.pdf")
            with open(pdf_path, "wb") as pdf_file:
                for chunk in pdf_response.iter_content(chunk_size=1024):
                    pdf_file.write(chunk)

            print(f"Downloaded: {file_name}.pdf")
        else:
            print(f"PDF not found for DOI: {doi}")

    except requests.exceptions.RequestException as e:
        print(f"Error fetching {doi}: {e}")
```

```python
for idx, row in df.iterrows():
    doi = row["RetractionDOI"]
    if pd.notna(doi):  # Ensure DOI is not NaN
        download_pdf(doi, idx + 1)  # Name file based on row index + 1
```

```python
import os
import time
import requests
from urllib.parse import urljoin
from bs4 import BeautifulSoup

# Directories
old_pdf_dir = "PDFs"  # Previous download folder
new_pdf_dir = "Pdf_attempt2"  # New folder for saving PDFs
os.makedirs(new_pdf_dir, exist_ok=True)

# Headers to mimic a browser
HEADERS = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"
}

def download_pdf(doi, file_name):
    base_url = "https://doi.org/"
    doi_url = base_url + doi

    # Check if PDF already exists in old directory
    existing_pdf_path = os.path.join(old_pdf_dir, f"{file_name}.pdf")
    if os.path.exists(existing_pdf_path):
        print(f"Skipping {file_name}.pdf (already exists in PDFs folder)")
        return

    try:
        # Start a session to handle cookies and headers
        session = requests.Session()
        session.headers.update(HEADERS)

        # Get redirect URL
        response = session.get(doi_url, allow_redirects=True)
        response.raise_for_status()
        final_url = response.url  # Resolved DOI page

        # Fetch final webpage with session
        page = session.get(final_url)
        page.raise_for_status()

        # Parse HTML
        soup = BeautifulSoup(page.text, "html.parser")

        # Look for meta citation PDF link
        pdf_link = None
        meta_tag = soup.find("meta", {"name": "citation_pdf_url"})
        if meta_tag:
            pdf_link = meta_tag["content"]
        
        # If no meta PDF link, search for PDF in anchor tags
        if not pdf_link:
            for link in soup.find_all("a", href=True):
                href = link["href"]
                if "pdf" in href.lower():
                    pdf_link = urljoin(final_url, href)
                    break  # Stop after first valid link

        # Attempt to download if PDF link found
        if pdf_link:
            pdf_response = session.get(pdf_link, stream=True)
            pdf_response.raise_for_status()

            # Save PDF in the new directory
            pdf_path = os.path.join(new_pdf_dir, f"{file_name}.pdf")
            with open(pdf_path, "wb") as pdf_file:
                for chunk in pdf_response.iter_content(chunk_size=1024):
                    pdf_file.write(chunk)

            print(f"Downloaded: {file_name}.pdf")
        else:
            print(f"PDF not found for DOI: {doi}")

    except requests.exceptions.RequestException as e:
        print(f"Error fetching {doi}: {e}")

    finally:
        time.sleep(2)  # Delay between requests to prevent blocking

```

```python
import os
import time
import requests
from urllib.parse import urljoin
from bs4 import BeautifulSoup

# Define directories
old_pdf_dirs = ["PDFs", "Pdf_attempt2"]  # Folders to check for existing PDFs
new_pdf_dir = "Pdf_attempt3"  # Folder to save new downloads
os.makedirs(new_pdf_dir, exist_ok=True)  # Create new folder if not exists

# Headers to mimic a browser and bypass 403 errors
HEADERS = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36",
    "Referer": "https://scholar.google.com",  # Helps bypass some site restrictions
    "Accept-Language": "en-US,en;q=0.9",
    "Connection": "keep-alive"
}

def pdf_already_exists(file_name):
    """Check if the PDF exists in any of the old directories."""
    for folder in old_pdf_dirs:
        if os.path.exists(os.path.join(folder, f"{file_name}.pdf")):
            return True
    return False

def extract_pdf_link(final_url, session):
    """Try different methods to extract a PDF link from the webpage."""
    try:
        page = session.get(final_url, headers=HEADERS)
        page.raise_for_status()
        soup = BeautifulSoup(page.text, "html.parser")

        # 1. Check for citation PDF meta tag
        meta_tag = soup.find("meta", {"name": "citation_pdf_url"})
        if meta_tag:
            return meta_tag["content"]

        # 2. Search for PDF in <a> tags
        for link in soup.find_all("a", href=True):
            href = link["href"]
            if "pdf" in href.lower() and "download" in href.lower():
                return urljoin(final_url, href)  # Resolve relative URLs

        # 3. Search for direct PDF links inside embedded <iframe> (some journals do this)
        iframe = soup.find("iframe", {"src": True})
        if iframe and "pdf" in iframe["src"].lower():
            return urljoin(final_url, iframe["src"])

    except requests.exceptions.RequestException as e:
        print(f"Error fetching page content: {e}")

    return None  # No PDF link found

def download_pdf(doi, file_name):
    """Download PDF from DOI link, handling restrictions & bypassing errors."""
    # Skip if file already exists in old directories
    if pdf_already_exists(file_name):
        print(f"Skipping {file_name}.pdf (already exists in PDFs or Pdf_attempt2)")
        return

    base_url = "https://doi.org/"
    doi_url = base_url + doi

    try:
        # Start a session to maintain cookies & headers
        session = requests.Session()
        session.headers.update(HEADERS)

        # Get redirect URL
        response = session.get(doi_url, allow_redirects=True)
        response.raise_for_status()
        final_url = response.url  # Resolved DOI page

        # Extract PDF link using multiple strategies
        pdf_link = extract_pdf_link(final_url, session)

        # Attempt to download the PDF if found
        if pdf_link:
            try:
                pdf_response = session.get(pdf_link, stream=True, headers=HEADERS)
                pdf_response.raise_for_status()

                # Save the PDF in the new folder
                pdf_path = os.path.join(new_pdf_dir, f"{file_name}.pdf")
                with open(pdf_path, "wb") as pdf_file:
                    for chunk in pdf_response.iter_content(chunk_size=1024):
                        pdf_file.write(chunk)

                print(f"Downloaded: {file_name}.pdf")

            except requests.exceptions.RequestException as e:
                print(f"Error downloading PDF for {doi}: {e}")

        else:
            print(f"PDF not found for DOI: {doi}")

    except requests.exceptions.RequestException as e:
        print(f"Error fetching {doi}: {e}")

    finally:
        time.sleep(2)  # Prevents rate-limiting by adding a short delay

```

```python
for idx, row in df.iterrows():
    doi = row["RetractionDOI"]
    if pd.notna(doi):  # Ensure DOI is not NaN
        download_pdf(doi, idx + 1)  # Name file based on row index + 1
```

```python
import os
import time
import requests
import pandas as pd
from bs4 import BeautifulSoup
from fpdf import FPDF

# Load your DataFrame (Assuming df is already loaded elsewhere in your script)
# df = pd.read_csv("your_file.csv")  # Uncomment and modify if needed

# Directories
existing_pdf_dirs = ["PDFs", "Pdf_attempt2", "Pdf_attempt3"]
new_pdf_dir = "Pdf_attempt4"
os.makedirs(new_pdf_dir, exist_ok=True)

# Headers to mimic a browser
HEADERS = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36",
    "Referer": "https://scholar.google.com",
    "Accept-Language": "en-US,en;q=0.9",
    "Connection": "keep-alive"
}

def pdf_exists(file_name):
    """Check if the PDF already exists in the specified directories."""
    for directory in existing_pdf_dirs:
        if os.path.isfile(os.path.join(directory, f"{file_name}.pdf")):
            return True
    return False

def save_text_as_pdf(text, file_path):
    """Save the extracted text as a PDF."""
    pdf = FPDF()
    pdf.add_page()
    pdf.set_auto_page_break(auto=True, margin=15)
    pdf.set_font("Arial", size=12)
    pdf.multi_cell(0, 10, text)
    pdf.output(file_path)

def extract_and_save_article(doi, file_name):
    """Extract the article content and save it as a PDF."""
    if pdf_exists(file_name):
        print(f"Skipping {file_name}.pdf (already exists).")
        return

    doi_url = f"https://doi.org/{doi}"
    try:
        response = requests.get(doi_url, headers=HEADERS, allow_redirects=True)
        response.raise_for_status()
        final_url = response.url

        page = requests.get(final_url, headers=HEADERS)
        page.raise_for_status()
        soup = BeautifulSoup(page.content, "html.parser")

        # Find the div with class 'article-section__content'
        content_div = soup.find("div", class_="article-section__content")
        if content_div:
            paragraphs = content_div.find_all("p")
            article_text = "\n\n".join(p.get_text() for p in paragraphs)
            if article_text.strip():
                pdf_path = os.path.join(new_pdf_dir, f"{file_name}.pdf")
                save_text_as_pdf(article_text, pdf_path)
                print(f"Extracted and saved as PDF: {file_name}.pdf")
            else:
                print(f"No extractable text found for DOI: {doi}")
        else:
            print(f"No content found with class 'article-section__content' for DOI: {doi}")

    except requests.RequestException as e:
        print(f"Error fetching {doi}: {e}")

def download_pdf(doi, file_name):
    """Attempt to download the PDF file from the DOI link."""
    if pdf_exists(file_name):
        print(f"Skipping {file_name}.pdf (already exists).")
        return

    base_url = "https://doi.org/"
    doi_url = base_url + doi

    try:
        # Start a session to maintain cookies & headers
        session = requests.Session()
        session.headers.update(HEADERS)

        # Get redirect URL
        response = session.get(doi_url, allow_redirects=True)
        response.raise_for_status()
        final_url = response.url  # Resolved DOI page

        # Extract potential PDF link
        soup = BeautifulSoup(session.get(final_url).text, "html.parser")
        pdf_link = None

        # Check for citation_pdf_url meta tag
        meta_tag = soup.find("meta", {"name": "citation_pdf_url"})
        if meta_tag:
            pdf_link = meta_tag["content"]

        # Search for PDF links in <a> tags
        if not pdf_link:
            for link in soup.find_all("a", href=True):
                href = link["href"]
                if "pdf" in href.lower():
                    pdf_link = href
                    break

        # Download the PDF if found
        if pdf_link:
            pdf_response = session.get(pdf_link, headers=HEADERS, stream=True)
            pdf_response.raise_for_status()

            pdf_path = os.path.join(new_pdf_dir, f"{file_name}.pdf")
            with open(pdf_path, "wb") as pdf_file:
                for chunk in pdf_response.iter_content(chunk_size=1024):
                    pdf_file.write(chunk)

            print(f"Downloaded: {file_name}.pdf")
        else:
            print(f"PDF not found for DOI: {doi}")
            extract_and_save_article(doi, file_name)  # Extract and save text as PDF

    except requests.exceptions.RequestException as e:
        print(f"Error fetching {doi}: {e}")

    finally:
        time.sleep(2)  # Prevents rate-limiting by adding a short delay

# **Process the DataFrame** - Loop through each DOI and try to download or extract
for idx, row in df.iterrows():
    doi = row["RetractionDOI"]
    if pd.notna(doi):  # Ensure DOI is not NaN
        download_pdf(doi, idx + 1)  # Name file based on row index + 1
```
